```python
import matplotlib.image as mpimg
import torch
from einops import rearrange
import matplotlib.pyplot as plt
```

```python
# tensor of 4x4 with each 2 rows 1 and 0s, first two rows are 1s

x = torch.zeros(16,4,3)
x[::2] = 1

plt.imshow(x, cmap='gray')
```

```python
import math

x_ = rearrange(x, '(h s1) (w s2) c -> h w (s2 s1 c)', s1=2, s2=2, h=8, w=2)
print(x_.shape)

fig, axs = plt.subplots(1,12, figsize=(12,8))
for i in range(12):
    axs[i].imshow(x_.numpy()[:,:,i], cmap='gray', vmin=0, vmax=1)
    axs[i].set_title(i)
    axs[i].axis('off')
```

```python
x0 = x[0::2, 0::2, :]  # B H/2 W/2 C
x1 = x[1::2, 0::2, :]  # B H/2 W/2 C
x2 = x[0::2, 1::2, :]  # B H/2 W/2 C
x3 = x[1::2, 1::2, :]  # B H/2 W/2 C

_x = torch.cat([x0, x1, x2, x3], -1)  # B H/2 W/2 4*C

fig, axs = plt.subplots(1,12, figsize=(12,8))
for i in range(12):
    axs[i].imshow(_x.numpy()[:,:,i], cmap='gray', vmin=0, vmax=1)
    axs[i].set_title(i)
    axs[i].axis('off')
```
